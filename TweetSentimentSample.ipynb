```python
import pandas as pd
import numpy as np
import preprocessor as p
from string import punctuation
import time
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
import matplotlib.pyplot as plt
```

```python
tweets = pd.read_csv("main_players2015.zip")
tweets = tweets.sample(n = 1000)
```

```python
def add_year(df, year):
    df['nfl_year'] = year
```

```python
add_year(tweets, 2015)
```

```python
tweets.drop_duplicates(inplace = True)
```

```python
tweets.drop(columns = ["Unnamed: 0"], inplace = True)
```

```python
tweets['clean_text'] = [p.clean(tweet).strip(punctuation) for tweet in tweets['text'].astype(str).tolist()]
```

```python
analyser = SentimentIntensityAnalyzer()
```

```python
tweets['neg'] = 0.0
tweets['neu'] = 0.0
tweets['pos'] = 0.0
tweets['compound'] = 0.0
tweets['sentiment'] = ''
for i, tweet in tweets.iterrows():
    score = analyser.polarity_scores(tweet.clean_text)
    tweets.at[i, 'neg'] = score['neg']
    tweets.at[i, 'neu'] = score['neu']
    tweets.at[i, 'pos'] = score['pos']
    tweets.at[i, 'compound'] = score['compound']
    if score['compound'] >= 0.05:
        tweets.at[i, 'sentiment'] = 'positive'
    elif score['compound'] <= -0.05:
        tweets.at[i, 'sentiment'] = 'negative'
    else:
        tweets.at[i, 'sentiment'] = 'neutral'
```
