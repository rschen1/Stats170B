```python
import pandas as pd
import numpy as np
```

```python
advanced_defense = pd.read_csv("game data/advanced_defense.csv")
advanced_offense = pd.read_csv("game data/advanced_offense.csv")
main_players = pd.read_csv("game data/mainplayers.csv")
pass_defense = pd.read_csv("game data/pass_defense.csv")
pass_offense = pd.read_csv("game data/pass_offense.csv")
rush_defense = pd.read_csv("game data/rush_defense.csv")
rush_offense = pd.read_csv("game data/rush_offense.csv")
```

```python
rush = pd.merge(rush_offense, rush_defense, on = ['team', 'nfl_year'], suffixes = ('_offense', '_defense'))
advanced = pd.merge(advanced_offense, advanced_defense, on = ['team', 'nfl_year'], suffixes = ('_offense', '_defense'))
pass_ = pd.merge(pass_offense, pass_defense, on = ['team', 'nfl_year'], suffixes = ('_offense', '_defense'))
```

```python
merge_cols = ['rush_tds_allowed',
 'pass_yds',
 'rush_yds',
 'pass_completed_offense',
 'nfl_year',
 'pass_tds_allowed',
 'pass_completed_defense',
 'pass_attempted_defense',
 'pass_yds_allowed',
 'rush_yds_allowed',
 'pass_tds',
 'rush_yds_per_attempt',
 'pass_attempted_offense',
 'rush_tds',
 'team']
merged = pd.merge(rush, pass_, on = ['team', 'nfl_year'], suffixes = ('_rush', '_pass'))
merged = pd.merge(merged, advanced, on = merge_cols)
#merged = pd.merge(merged, advanced, on = ['team', 'nfl_year'], suffixes = ('', '_advanced'))
```

```python
main_players.dropna(inplace = True)
```

```python
player_totals = main_players.groupby(['player_name', 'nfl_year', 'player_team']).agg('sum').reset_index().drop(['game_number', 'player_year'], axis = 1)
```

```python
player_totals = player_totals.add_suffix("_player")
```

```python
players = pd.merge(merged, player_totals, 
                   right_on = ['nfl_year_player', 'player_team_player'], 
                   left_on = ['nfl_year', 'team'], 
                   left_index = True, 
                   suffixes = ('_player', '_team'))
```
